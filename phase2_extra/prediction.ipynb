TRANSFORMATION - EQUIWIDTH

~~~~~~~~~~~~~~~~~~~

```python
for con_feature in continuous_features:
    # get the range of the descriptive feature:
    max = df[con_feature].max()
    min = df[con_feature].min()

    # define the number of bins:
    numOfBins = 10

    # define the bin width:
    width = (max-min)/numOfBins

    # create a histogram using equi-width:
    sns.histplot(data=df, x=con_feature, binwidth=width)

    # display histogram
    plt.show()
    
```

```python
# traverse the collection of continuous features:
for con_feature in continuous_features:

    # define the total number of instances:
    total_values = df[con_feature].value_counts().sum()

    # define the number of bins:
    numOfBins = 10

    # calculate the frequency:
    frequency = total_values / numOfBins
    
    # create a histogram using equi-frequency:
    sns.histplot(data=df, x=con_feature, stat='density', bins=numOfBins)

    # display histogram
    plt.show()
```
